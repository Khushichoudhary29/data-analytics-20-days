```python
# Student Grade System
marks = int(input("Enter marks (0‚Äì100): "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
elif marks >= 40:
    print("Grade D")
else:
    print("Fail")
```

```python
# Simple Contact Book
contact = {}

while True:
    print("\n1. Add Contact")
    print("2. Search")
    print("3. Show All")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        name = input("Enter name: ")
        phone = input("Enter phone: ")
        contact[name] = phone
        print("Contact added!")

    elif choice == 2:
        name = input("Enter name to search: ")
        print("Phone:", contact.get(name, "Not found"))

    elif choice == 3:
        for name, phone in contact.items():
            print(name, ":", phone)

    elif choice == 4:
        break

    else:
        print("Invalid choice")
```

#Ô∏è‚É£ üìå SECTION 3: SOLUTIONS OF 20 PRACTICE QUESTIONS



```python
# 1. Swap two numbers
a, b = 5, 10
a, b = b, a
print(a, b)

# 2. Largest of three numbers
a, b, c = 10, 50, 30
print(max(a, b, c))

# 3. Print numbers 1 to 100
for i in range(1, 101):
    print(i)

# 4. Sum of first 50 numbers
print(sum(range(1, 51)))

# 5. Check prime
num = 29
flag = True
for i in range(2, int(num**0.5)+1):
    if num % i == 0:
        flag = False
        break
print("Prime" if flag else "Not Prime")

# 6. Print table
n = 6
for i in range(1, 11):
    print(n, "x", i, "=", n*i)

# 7. Count vowels
s = "Khushi Choudhary"
count = sum(1 for ch in s.lower() if ch in 'aeiou')
print(count)

# 8. Reverse a number
num = 12345
rev = int(str(num)[::-1])
print(rev)

# 9. Reverse a string
print("Hello"[::-1])

# 10. Check palindrome
s = "madam"
print("Palindrome" if s == s[::-1] else "Not")

# 11. Factorial
fact = 1
n = 5
for i in range(1, n+1):
    fact *= i
print(fact)

# 12. Fibonacci
a, b = 0, 1
for _ in range(10):
    print(a, end=" ")
    a, b = b, a+b

# 13. Count words
s = "I love data analytics"
print(len(s.split()))

# 14. Max in list
lst = [3, 9, 2, 11, 7]
print(max(lst))

# 15. Remove duplicates
lst = [1,2,2,3,4,4,5]
print(list(set(lst)))

# 16. Armstrong number
num = 153
s = sum(int(d)**3 for d in str(num))
print("Armstrong" if s == num else "Not")

# 17. Sum of digits
num = 1234
total = sum(int(d) for d in str(num))
print(total)

# 18. Character frequency
s = "banana"
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
print(freq)

# 19. Merge lists (without +)
a = [1,2,3]
b = [4,5,6]
for x in b:
    a.append(x)
print(a)

# 20. Print pattern
for i in range(1, 6):
    print("*" * i)
```
