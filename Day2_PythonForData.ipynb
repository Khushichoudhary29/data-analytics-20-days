```python
# Q1: Count lines
with open("sample.txt", "r") as f:
    print(len(f.readlines()))

# Q2: Remove duplicates from string
s = "banana"
result = "".join(sorted(set(s)))
print(result)

# Q3: List of strings ‚Üí integers
lst = ["1","2","3"]
print(list(map(int, lst)))

# Q4: Dictionary from two lists
keys = ["name", "age"]
values = ["Khushi", 20]
print(dict(zip(keys, values)))

# Q5: Count upper/lower
s = "Hello Khushi"
u = sum(1 for ch in s if ch.isupper())
l = sum(1 for ch in s if ch.islower())
print(u, l)

# Q6: 2nd largest
lst = [10, 50, 20, 40]
print(sorted(lst)[-2])

# Q7: Sort dict by values
d = {"a":3, "b":1, "c":2}
print(dict(sorted(d.items(), key=lambda x: x[1])))

# Q8: Reverse each word
s = "I love data"
print(" ".join(word[::-1] for word in s.split()))

# Q9: Squares 1‚Äì20
print([i*i for i in range(1,21)])

# Q10: Common elements
a = [1,2,3]
b = [2,3,4]
print(list(set(a) & set(b)))

# Q11: Check subset
print(set([1,2]).issubset([1,2,3,4]))

# Q12: Words starting with vowels
words = ["apple","cat","orange","dog"]
print([w for w in words if w[0].lower() in 'aeiou'])

# Q13: Replace vowels
s = "Khushi"
for v in 'aeiouAEIOU':
    s = s.replace(v, '*')
print(s)

# Q14: Flatten nested list
nested = [[1,2],[3,4],[5]]
flat = [x for sub in nested for x in sub]
print(flat)

# Q15: Remove None
lst = [1, None, 2, None, 3]
print([x for x in lst if x is not None])

# Q16: Merge dictionaries
d1 = {"a":1}
d2 = {"b":2}
d = {**d1, **d2}
print(d)

# Q17: Word frequency
s = "I Love Data I Love Python"
words = s.split()
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1
print(freq)

# Q18: Extract digits
s = "a1b2c3"
digits = [int(ch) for ch in s if ch.isdigit()]
print(digits)

# Q19: Sum of numbers in string
s = "a1b2c3"
total = sum(int(ch) for ch in s if ch.isdigit())
print(total)

# Q20: Person class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(self.name, self.age)

p = Person("Khushi", 20)
p.show()
```
